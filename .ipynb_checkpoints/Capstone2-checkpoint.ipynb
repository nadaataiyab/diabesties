```python
#note that .apply() returns a series, so have to feed it back to the dataframe
def alc_update(row):
    if row.action == 'a1c_update':
        return 'a1c_update'
    else:
        return row.object_name
```

```python
def convert_datetime(df, cols_datetime):
    '''
    Parameters
    ----------
    df: dataframe 
    cols_datetime: list of column names to convert to date time type
    
    Returns
    -------
    Dataframe with features changed to datetime objects
    
    '''
    
    df = df.copy()
    for col in cols_datetime:
        df[col] = pd.to_datetime(df[col])
    return df
```

```python
def convert_categories(df, cols_categories):
    '''
    Parameters
    ----------
    df: dataframe 
    cols_datetime: list of column names to convert to category type
    
    Returns
    -------
    Dataframe with features changed to datetime objects
    
    '''
    
    df = df.copy()
    for col in cols_categories:
        df[col] = df[col].astype('category')
    return df
```

```python
df_dem_7 = convert_categories(df_dem_7, ['diabetes_type', 'ethnicity', 'gender'])
```

```python
df_dem_7.info()
```

```python
 def dummy_variables(df):
    df = df.copy()
    dummy_cols = df.select_dtypes(include=['category']).columns
    for col in dummy_cols:
        dummies = pd.get_dummies(df[col], dummy_na=True, drop_first=True, prefix=col)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(col, axis=1)
    return df
```

```python
df_dem_7 = dummy_variables(df_dem_7)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import roc_curve, auc, confusion_matrix, precision_score, recall_score 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
```

# Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
y_score = lr.predict_proba(X_test)[:, 1]
y_true = y_test
fpr, tpr, thresholds = roc_curve(y_true, y_score)
plt.plot(fpr, tpr)
print(auc(fpr, tpr))
```

```python
y_pred = lr.predict(X_test)
print(precision_score(y_true, y_pred))
print(recall_score(y_true, y_pred))
```
